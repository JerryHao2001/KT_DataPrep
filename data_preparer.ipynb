```python
import pandas as pd
import json
```

```python
def process_questions(path):
    # Load the Excel file
    xls = pd.ExcelFile(path)

    # Initialize an empty list to hold DataFrames
    dataframes = []

    # Iterate over each sheet in the Excel file
    for sheet_name in xls.sheet_names:
        # Read the sheet into a DataFrame
        df = pd.read_excel(xls, sheet_name=sheet_name)
        
        # Keep columns L, M, and Q (which correspond to columns 11, 12, and 16 in 0-based indexing)
        # columns_to_keep = df.iloc[:, [11, 12, 16]]
        
        # Append the DataFrame to the list
        dataframes.append(df)

    # Concatenate all DataFrames into a single DataFrame
    questions = pd.concat(dataframes, ignore_index=True)
    questions = questions.loc[:, ["类型", "题型", "全站正确率"]]
    questions = questions.dropna()  
    questions['类型_题型'] = questions['类型'] + '_' + questions['题型']
    questions = questions.drop(columns=['类型', '题型'])

    concept_dict = {i:j for i,j in zip(questions['类型_题型'].unique(), range(len(questions['类型_题型'].unique())))}
    questions['类型_题型'] = questions['类型_题型'].map(concept_dict)

    questions = questions.rename(columns={'类型_题型': 'concept_id', '全站正确率': 'difficulty'})
    questions['difficulty'] = questions['difficulty'].apply(lambda x: 1-x)
    questions['question_id'] = questions.index
    questions = questions[['question_id','concept_id','difficulty']]

    questions.to_csv('cleand_question.csv', index=False)
    with open('concept_dict.json', 'w') as file:
        json.dump(concept_dict, file, indent=4)


    return questions, concept_dict
```

```python
file_path = 'Questions_46.xlsx'
process_questions(file_path)
```
